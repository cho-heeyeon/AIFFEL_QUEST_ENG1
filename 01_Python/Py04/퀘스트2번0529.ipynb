# 퀘스트 2번: 간단한 자동사냥 RPG 만들기

```python
import random
 
 
class Character:
    def __init__(self, name: str, level: int, hp: int, atk: int, defense: int):
        self.name = name
        self.level = level
        self.hp = hp
        self.atk = atk
        self.defense = defense
 
    def is_alive(self) -> bool:
        return self.hp > 0
 
    def take_damage(self, damage: int) -> int:
        real_damage = damage - self.defense
        if real_damage < 0:
            real_damage = 0
 
        self.hp -= real_damage
        return real_damage
 
    def attack_target(self, target: "Character") -> int:
        damage = random.randint(1, self.atk)
        target.take_damage(damage)
        return damage
 
 
class Player(Character):
    def __init__(self, name: str):
        super().__init__(name=name, level=1, hp=100, atk=25, defense=5)
        self.exp = 0
 
    def gain_experience(self, amount: int) -> None:
        self.exp += amount
 
    def level_up(self) -> None:
        while self.exp >= 50:
            self.exp -= 50
            self.level += 1
            self.atk += 10
            self.defense += 5
 
 
class Monster(Character):
    def __init__(self, name: str, level: int):
        hp = random.randint(10, 30) * level
        atk = random.randint(5, 20) * level
        defense = random.randint(1, 5) * level
        super().__init__(name=name, level=level, hp=hp, atk=atk, defense=defense)
 
 
def battle(player: Player, monster: Monster) -> bool:
    print(f"\n{monster.name} 과의 전투를 시작합니다.")
 
    while player.is_alive() and monster.is_alive():
        dmg = player.attack_target(monster)
        print(f"{player.name}이 {monster.name}에게 {dmg} 만큼 공격했다...!")
        print(f"{monster.name}의 체력: {monster.hp}")
 
        if not monster.is_alive():
            break
 
        dmg = monster.attack_target(player)
        print(f"{monster.name}이 {player.name}에게 {dmg} 만큼 공격했다...!")
        print(f"{player.name}의 체력: {player.hp}")
 
    if player.is_alive():
        gained = monster.level * 20
        player.gain_experience(gained)
        player.level_up()
        print("전투 승리!")
        print(f"{monster.name} 을 이겼다!")
        return True
 
    print("전투 패배..")
    return False
 
 
def main() -> None:
    monster_dict = {"슬라임": 1, "고블린": 2, "오크": 3}
 
    name = input("플레이어 이름을 입력하세요: ").strip()
    if not name:
        name = "히어로"
 
    player = Player(name)
    print(f"\n{name} 님, 자동사냥을 시작합니다!")
    print(f"플레이어 상태: 레벨 {player.level}, 체력 {player.hp}, 공격력 {player.atk}, 방어력 {player.defense}, 경험치 {player.exp}")
 
    for m_name, m_level in monster_dict.items():
        monster = Monster(m_name, m_level)
        win = battle(player, monster)
 
        if not win:
            print("\n게임오버")
            return
 
        print(f"현재 상태: 레벨 {player.level}, 체력 {player.hp}, 공격력 {player.atk}, 방어력 {player.defense}, 경험치 {player.exp}")
 
    print("\n모든 몬스터를 처치했습니다! 게임 클리어!")
 
 
main()
```
